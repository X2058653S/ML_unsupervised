### K-MEANS

Clasificamos los datos con el algoritmo K-means. Buscamos el número óptimo de clusters :

### ELBOW METHOD

```python
kmeans_kwargs={'init':'random', 'n_init':20,'max_iter':500,'random_state':42}

# A list holds the SSE values for each k
sse = []
for k in range(1,11):
    kmeans = KMeans(n_clusters=k, **kmeans_kwargs)
    kmeans.fit(df_cluster_final)
    sse.append(kmeans.inertia_)
```

```python
sse
```

```python
plt.plot(range(1,11), sse)
plt.xticks(range(1,11))
plt.xlabel("Number of Clusters")
plt.ylabel("SSE")
plt.show()
```

```python
kl = KneeLocator(range(1,11), sse, curve='convex', direction='decreasing')
kl.elbow
```

En un principio, y aunque el cálculo nos devuelva un resultado de 4, parece bastante claro según la gráfica que el codo se sitúa en el número de 2 clusters. Reconfirmamos el dato con silhouette :

### SILHOUETTE

```python
# A list holds the silhouette coefficients for each k
silhouette_coefficients = []

# Notice you start at 2 clusters for silhouette coefficient
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, **kmeans_kwargs)
    kmeans.fit(df_cluster_final)
    score = silhouette_score(df_cluster_final, kmeans.labels_)
    silhouette_coefficients.append(score)
```

```python
silhouette_coefficients
```

```python
plt.plot(range(2, 11), silhouette_coefficients)
plt.xticks(range(2, 11))
plt.xlabel("Number of Clusters")
plt.ylabel("Silhouette Coefficient")
plt.show()
```

El mejor resultado se obtiene con 2 clusters.

```python
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage='ward')  
cluster.fit_predict(df_cluster_final)
```

```python
df_labels_agg=pd.DataFrame(cluster.fit_predict(df_cluster_final),columns=['Label_agg'])
df_labels_agg
```

```python
df_labels_agg.Label_agg.value_counts()
```

El modelo nos propone 2 clusters, uno con 560 registros y el otro con el resto.

```python
df_agg_final=pd.concat([df,df_labels_agg], axis=1)
```

```python
df_agg_final[df_agg_final['Label_agg']==1].describe()
```

```python
df_agg_final[df_agg_final['Label_agg']==0].describe()
```

Con este algoritmo también, todo indica que el cluster con menos registros agrupa las tops publicaciones en cuanto a reaaciones, comentarios y shares.

```python
# A list holds the silhouette coefficients for each k
silhouette_coefficients_agg = []

# Notice you start at 2 clusters for silhouette coefficient
for k in range(2, 11):
    cluster = AgglomerativeClustering(n_clusters=k, affinity='euclidean', linkage='ward')  
    cluster.fit_predict(df_cluster_final)
    score = silhouette_score(df_cluster_final,cluster.fit_predict(df_cluster_final))
    silhouette_coefficients_agg.append(score)

plt.plot(range(2, 11), silhouette_coefficients_agg)
plt.xticks(range(2, 11))
plt.xlabel("Number of Clusters")
plt.ylabel("Silhouette Coefficient")
plt.show()
```

```python
silhouette_coefficients_agg
```

El análisis silhouette con 2 clusters nos devuelve un resultado de 0,76 con 2 clusters, algo menos que K-means que tiene un resultado de 0,85 el cual es por lo tanto el mejor de los dos modelos.
